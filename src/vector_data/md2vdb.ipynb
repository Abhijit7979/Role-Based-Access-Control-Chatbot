```python
from langchain_text_splitters import MarkdownHeaderTextSplitter,RecursiveCharacterTextSplitter

headers_to_split_on = [
    ("#", "Header 1"),
    ("##", "Header 2"),
    ("###", "Header 3"),
]
```

```python
markdown_splitter = MarkdownHeaderTextSplitter(headers_to_split_on,strip_headers=False)
recursive_splitter = RecursiveCharacterTextSplitter(chunk_size=1000, chunk_overlap=200)
```

```python
with open("../../data/engineering/engineering_master_doc.md", "r", encoding="utf-8") as file:
    md_content = file.read()
chunks=markdown_splitter.split_text(md_content)
```

```python
print(chunks)
```

```python
from langchain_community.embeddings import OpenAIEmbeddings 
from langchain_chroma import Chroma
```

```python
embeddings = OpenAIEmbeddings(model="text-embedding-3-large")
db=Chroma.from_documents(documents=chunks,embedding=embeddings,persist_directory="Engineering_doc")
```

```python
print(db.similarity_search("10.3 Contact Information"))
```
